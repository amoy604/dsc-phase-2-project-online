```python
target = 'price'
columns_correlations = []
columns_non_numeric = []

for column in df.drop(columns=[target]).columns:
    try:
        corr = np.abs(df[column].corr(df[target]))
        t = (column, corr)
        columns_correlations.append(t)
    except:
        columns_non_numeric.append(column)
```

```python
columns_correlations = sorted(columns_correlations, key = lambda x: x[1], reverse=True)
columns_correlations
```

```python
correlated_features_above_20 = [t[0] for t in columns_correlations if t[1] >= 0.20]
correlated_features_above_20
```

```python
model_df = correlated_features_above_20
model_df.append(target)
```

```python
correlated_df = df[model_df]
```
